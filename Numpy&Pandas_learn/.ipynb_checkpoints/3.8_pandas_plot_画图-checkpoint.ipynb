```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Series
data=pd.Series(np.random.randn(1000),index=np.arange(1000))
data=data.cumsum()
data.plot()
#显示
#横坐标：index，纵坐标：value
plt.show()
```

```python
#DataFrame
data2=pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000),columns=list('ABCD'))
data2=data2.cumsum()
print(data2.head())
data2.plot()
#4组数据对应每一列
plt.show()
```

```python
#plt method
#plot-bar-hist-box-kde-area-scatter-hexbin-pie
ax=data2.plot.scatter(x='A',y='B',color='DarkBlue',label='Class 1')
#ax=ax:将上面的点添加到此图上
data2.plot.scatter(x='A',y='C',color='DarkGreen',label='Class 2',ax=ax)

plt.show()
```
