```python
import numpy as np
```

```python
#x,y,z轴坐标范围=0:3,0:3,0:3
a=np.mgrid[0:3,0:3,0:3]
print(a.shape)
print(a)
#获取网格中所有坐标点位置
x=a.reshape(3,-1)
y=x.T
print(y)
```

```python
start=np.array((-12,-12,52))
stop=np.array((144,144,144))+start
step=np.array((64,64,64))
b=np.mgrid[start[0]:stop[0]:step[0], start[1]:stop[1]:step[1],
                               start[2]:stop[2]:step[2]]
```

```python
print(b.shape)
print(b)
```

```python
c=b.reshape(3,-1)#每行的值分别为x/y/z轴索引取值
d=c.T#转置后，每行为1个坐标位置
print(c.shape)
print(c)
print(d.shape)
print(d)
```
