## Python处理csv文件

CSV(Comma-Separated Values)即逗号分隔值，可以用Excel打开查看。由于是纯文本，任何编辑器也都可打开。与Excel文件不同，CSV文件中：

- 值没有类型，所有值都是字符串
- 不能指定字体颜色等样式
- 不能指定单元格的宽高，不能合并单元格
- 没有多个工作表
- 不能嵌入图像图表

在CSV文件中，以,作为分隔符，分隔两个单元格。像这样``a,,c``表示单元格``a``和单元格``c``之间有个空白的单元格。依此类推。

不是每个逗号都表示单元格之间的分界。所以即使CSV是纯文本文件，也坚持使用专门的模块进行处理。Python内置了csv模块。先看看一个简单的例子。

### 1.从CSV文件中读取数据

注意：
- open函数指定encoding以文件保存的编码打开，默认utf-8
- ``data``不能直接打印，list(data)最外层是list，**里层的每一行数据都在一个list中**

```python
import csv
filename1="./result1.csv"
filename2="./result2.csv"
```

#### 读取所有内容

```python
with open(filename1,encoding="utf-16") as f:
    reader=csv.reader(f)
    print(list(reader))
```

#### 逐行读取内容

```python
with open(filename1,encoding="utf-16") as f:
    reader=csv.reader(f)
    for row in reader:
        #打印每行行号和行内容
        print(reader.line_num,row)
```

### example1_size-op/s:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_lst=[]
y1_ssd=[]
y2_hdd=[]
y3_gffs=[]

with open(filename1,encoding="utf-16") as f:
    reader=csv.reader(f)
    
    #读取表头
    header_row=next(reader)
    print(header_row)
    for row in reader:
        #打印每行行号和行内容
        print(reader.line_num,row)
        x_lst.append(int(row[0]))
        y1_ssd.append(float(row[1]))
        y2_hdd.append(float(row[2]))
        y3_gffs.append(float(row[3]))
```

```python
print(type(x_lst[0]))
print(type(y1_ssd[0]))
print(type(y2_hdd[0]))
print(type(y3_gffs[0]))
```

```python
x=np.asarray(x_lst)
y1=np.asarray(y1_ssd)/10000
y2=np.asanyarray(y2_hdd)/10000
y3=np.asarray(y3_gffs)/10000
```

```python
plt.figure(figsize=(8,6))
x_ticks=np.linspace(0,20,6)
plt.xticks(x_ticks)
plt.xlabel("size(KB)")
plt.ylabel("op(10000)/s")

plt.plot(x,y1,color='red',label="ssd")
plt.scatter(x,y1,s=50,color="black")

plt.legend(loc="upper right")


plt.show()
```

```python
plt.figure(figsize=(8,6))
x_ticks=np.linspace(0,20,6)
plt.xticks(x_ticks)
plt.xlabel("size(KB)")
plt.ylabel("op(10000)/s")

plt.plot(x,y2,color='blue',label="hdd")
plt.scatter(x,y2,s=50,color="black")

plt.plot(x,y3,color='green',label="gffs-ssd+hdd")
plt.scatter(x,y3,s=50,color="black")

plt.legend(loc='upper right')

plt.show()
```
